```
from evaluation import heatmap_on_test
from google.colab import drive
import tensorflow as tf
import pandas as pd
import os
```

## Load Model

```
drive.mount('/content/gdrive/')
drive_root_dir = '/content/gdrive/MyDrive/'
model = tf.keras.models.load_model(drive_root_dir + 'level_3_tl_inception_scratch_data_aug.h5')
level = 4
slice_size = 299
```

## Heatmap on tumor_096.tif

```
cnf_matrix, precision, recall = heatmap_on_test(level,'tumor_096',slice_size,model,class_names=['0negative','1positive']).generate()
```

```
print(pd.DataFrame(cnf_matrix, columns=['Pred Negative','Pred Positive'], index=['Actual Negative','Actual Positive']))
print(f'\nprecision: {precision}')
print(f'recall: {recall}')
```
